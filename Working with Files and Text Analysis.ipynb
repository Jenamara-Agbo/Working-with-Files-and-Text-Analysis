### Read vs Readlines
Above, we see that we can grab all the content of the fie using read(). This is a great way to go if we want to work with the file content as a whole, for example, searching for a word in the entire file, or processing word frequency distribution on a presidential speech.

But what if the the file is very large? Teading the entire content into memory can be inefficient or even cause memory issues. 

Or what if we want to process the file line by line and pull out spaces, or other extraneous info that we don't want? 

Readlines to the rescue! 
Readlines reads the file line by line and returns a list of strings, where each string represents a line from the file.

Use readlines() when you need to work with the file line by line, and you want to process or iterate over the lines.

Here is an example:

```python
with open('example.txt', 'r') as file:
    lines = file.readlines()  # Reads the file and returns a list of lines
for line in lines:
    print(line.strip())  # Strip removes newline characters
```

```python
with open('first_lines_Kennedy_John.txt', 'r') as file:
    lines = file.readlines()  # Reads the file and returns a list of lines
for line in lines:
    print(line.strip())  # Strip removes newline characters
```

### Alternative for very large files: 
I want to introduce you to this idea, because ChatGPT might suggest it, so I want you to be exposed and know what's going on.
For very large files, neither read() nor readlines() may be efficient. Instead, you can read the file line by line without loading the whole file into memory using what is called a file object iterator:



```python

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

That said, you should use read() or readlines() in the majority of your cases as your file size will be moderately large at most.

```python
with open('first_lines_Kennedy_John.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
#Importing the needed packages
import nltk
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from nltk.corpus import inaugural 
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

#printing out all the fileids (speeches)
print(inaugural.fileids())  
```

### Chose a particular speech - by John F. Kennedy; then tokenized, and cleaned by removing stopwords

```python
#choosing a particular speech, tokenizing, cleaning by removing stopwords
kennedy_speech = inaugural.raw('1961-Kennedy.txt')  #John F. Kennedy's speech
token_kennedy_speech = word_tokenize(kennedy_speech)
#removing stopwords from token_kennedy'speech
stop_words = set(stopwords.words('english')) #the stopwords are enclosed in a set so there won't be any repetitin
cleaned_kennedy_speech = [] #the filtered speech after stopwords have been removed will be added to this list

#using a for loop to do the cleaning and filtering
for word in token_kennedy_speech:
    if word.isalpha() and word.lower() not in stop_words:
        cleaned_kennedy_speech.append(word)

print(cleaned_kennedy_speech) #this is now a list of cleaned words without any stop
```

### Created a frequency distribution of the common words

```python
#I will then created a frequency distribution of these words to see the most common words
fdist = FreqDist(cleaned_kennedy_speech)
print(fdist) #the result from this shows that there are 454 distinct words in the cleaned-kennedy_speech I created but 684 outcomes in total
#to display on the most common 30 words
print(fdist.most_common(30))

#visualizing these words
plt.figure(figsize=(10,6)) #this controls the size and the dimension of the plot
fdist.plot(30, cumulative=False) #this will display only the 30 most frequent words
plt.show()

#the result above shows that "US", "LET" have one of the highest frequencies, but they don'e really matter
```

### Looping through the speeches of five presidents at a time

```python
#Having written the code for just one president, I would create a loop to go through the speeches of five presidents at a time

def process_speeches(fileid):
    """
    This function will process each of the presidents' speeches and return cleaned tokens
    """
    speeches = inaugural.raw(fileid) #this will read in all the speeches
    tokenized_speeches = word_tokenize(speeches) #tolenizing the speeches
    #cleaning the tokenized speeches
    cleaned_tokenized_speeches = []
    for word in tokenized_speeches:
        if word.isalpha() and word.lower() not in stop_words:
            cleaned_tokenized_speeches.append(word)
    
    return cleaned_tokenized_speeches

#Manually selecting a few speeches from past presidents that I want to analyze
presidents = ['1789-Washington.txt', '1861-Lincoln.txt', '1905-Roosevelt.txt', '1953-Eisenhower.txt', '1961-Kennedy.txt', 
              '1997-Clinton.txt', '2001-Bush.txt', '2009-Obama.txt', '2017-Trump.txt', '2021-Biden.txt'] 

#I created a loop to go through the speeches
for president in presidents:
    print(f"Analyzing speech: {president}")
    #processing the speech
    tokens_president = process_speeches(president) #calling the function above to tokenize each speech once mentioned
    fdist = FreqDist(tokens_president) #created a frequency distribution of the cleaned words

    #Plotting the most common words 20 in each speech
    plt.figure(figsize=(10,6))
    #this splitting in the code below removes the year and .txt while printing the presidents' names.
    fdist.plot(30, cumulative=False, title=f"Most Common Words in {president.split('-')[1].split('.')[0]} Inaugural Speech")
    plt.show()
```

### Correlation between Clinton and Obama's speech during periods of economic recovery 

```python
""" According to history, Clinton and Obama for instance, delivered speeches during periods of economic recovery 
    (that is Clinton after the early 90's recession and Obama after the Great Recession in 2009. Thus, I analyzed 
    both of the separately to see if there are common economic themes in their speeches."""

# Speeches of interest: Clinton 1997 and Obama 2009 - put into a dictionary for easy access
presidents1 = {
    'Clinton': '1997-Clinton.txt',
    'Obama': '2009-Obama.txt'}

# Define theme-related words
economic_words = ['economy', 'jobs', 'growth', 'recovery', 'prosperity', 'market', 'recession']
technology_words = ['technology', 'innovation', 'globalization', 'internet', 'global', 'competition']
hope_words = ['hope', 'change', 'future', 'dream', 'opportunity']
inclusivity_words = ['unity', 'together', 'inclusive', 'we', 'community', 'equal', 'fair']

def process_speeches1(fileid):
    """This function will process each of these two presidents' speeches and return cleaned tokens"""
    speeches = inaugural.raw(fileid)
    token_word1 = word_tokenize(speeches)
    stop_words = set(stopwords.words('english'))
    cleaned_token_word = []
    for word in token_word1:
        if word.isalpha() and word.lower() not in stop_words:
            cleaned_token_word.append(word.lower())  # Converting to lowercase for consistency
    #returning the new list of the cleaned tokens
    return cleaned_token_word

# Function to count the frequency of theme-related words
def count_theme_words(token_word1, theme_words):
    fdist = FreqDist(token_word1)  # Corrected from `token_word` to `token_word1`
    theme_word_counts = {}  # using a dictionary to capture both the word and the number of times it occurred
    for word in theme_words:
        if word in fdist:
            theme_word_counts[word] = fdist[word]  # This will capture the number of times a word occurred
    return theme_word_counts

# Looping through each president's speech
for president, speech in presidents1.items():
    print(f"Analyzing {president}'s speech")
    
    # Process the speech
    token_word1 = process_speeches1(speech)  # Call the function
    
    # Count word frequencies for each theme
    economic_count = count_theme_words(token_word1, economic_words)
    technology_count = count_theme_words(token_word1, technology_words)
    hope_count = count_theme_words(token_word1, hope_words)
    inclusivity_count = count_theme_words(token_word1, inclusivity_words)
    
    # Print results
    print(f"\n{president}'s Economic-related words frequency: {economic_count}")
    print(f"{president}'s Technology/Globalization-related words frequency: {technology_count}")
    print(f"{president}'s Hope-related words frequency: {hope_count}")
    print(f"{president}'s Inclusivity-related words frequency: {inclusivity_count}")
    print("\n")
```

### Speech references to  religious themes (e.g., 'God', 'faith'), political ideals (e.g., 'freedom', 'democracy'), and socio-cultural topics (e.g., 'rights', 'equality')

```python
""" Religion, politics, and socio-cultural values have shaped American presidential rhetoric for decades. 
    This analysis looks at how different presidents have referenced religious themes (e.g., 'God', 'faith'), 
    political ideals (e.g., 'freedom', 'democracy'), and socio-cultural topics (e.g., 'rights', 'equality') 
    in their speeches to connect with the American public. """

# Speeches of interest - put into a dictionary for easy access
presidents = {
    'Kennedy': '1961-Kennedy.txt',
    'Bush': '2001-Bush.txt',
    'Obama': '2009-Obama.txt',
    'Trump': '2017-Trump.txt',
    'Biden': '2021-Biden.txt'
}

# Define theme-related words
religion_words = ['god', 'faith', 'blessing', 'pray', 'lord']
political_words = ['freedom', 'democracy', 'rights', 'justice', 'nation', 'liberty']
socio_cultural_words = ['equality', 'community', 'together', 'diversity', 'unity']

def process_speeches(fileid):
    """This function processes each president's speech and returns cleaned tokens."""
    speech = inaugural.raw(fileid)
    token_words = word_tokenize(speech)
    stop_words = set(stopwords.words('english'))
    cleaned_token_words = []
    for word in token_words:
        if word.isalpha() and word.lower() not in stop_words:
            cleaned_token_words.append(word.lower())  # Convert to lowercase for consistency
    return cleaned_token_words

# Function to count the frequency of theme-related words
def count_theme_words(tokens, theme_words):
    fdist = FreqDist(tokens)
    theme_word_counts = {}  # Dictionary to store word frequencies
    for word in theme_words:
        if word in fdist:
            theme_word_counts[word] = fdist[word]  # Capture the frequency of each word
    return theme_word_counts

# Loop through each president's speech
for president, speech in presidents.items():
    print(f"Analyzing {president}'s speech")
    
    # Process the speech
    tokens = process_speeches(speech)
    
    # Count word frequencies for each theme
    religion_count = count_theme_words(tokens, religion_words)
    political_count = count_theme_words(tokens, political_words)
    socio_cultural_count = count_theme_words(tokens, socio_cultural_words)
    
    # Print results
    print(f"\n{president}'s Religion-related words frequency: {religion_count}")
    print(f"{president}'s Political-related words frequency: {political_count}")
    print(f"{president}'s Socio-cultural-related words frequency: {socio_cultural_count}")
    print("\n")
```

### Frequency of religious, political, and socio-cultural themes

```python
# I visualized the visualize the frequency of religious, political, and socio-cultural themes in presidential 
#speeches using a bar chart through matplotlib

import matplotlib.pyplot as plt

# Function to plot word frequencies
def plot_word_frequencies(president, theme_counts, theme_name):
    """Plots the frequency of words for a given theme in a speech."""
    words = list(theme_counts.keys())
    counts = list(theme_counts.values())
    
    plt.figure(figsize=(10, 5))
    plt.bar(words, counts, width=0.3, color='skyblue')
    plt.title(f'{theme_name} Theme Word Frequencies in {president}\'s Speech')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.show()

# Loop through each president's speech and plot frequencies
for president, speech in presidents.items():
    print(f"Analyzing {president}'s speech")
    
    # Process the speech
    tokens = process_speeches(speech)
    
    # Count word frequencies for each theme
    religion_count = count_theme_words(tokens, religion_words)
    political_count = count_theme_words(tokens, political_words)
    socio_cultural_count = count_theme_words(tokens, socio_cultural_words)
    
    # Print results
    print(f"\n{president}'s Religion-related words frequency: {religion_count}")
    print(f"{president}'s Political-related words frequency: {political_count}")
    print(f"{president}'s Socio-cultural-related words frequency: {socio_cultural_count}")
    
    # Plot word frequencies for each theme
    plot_word_frequencies(president, religion_count, 'Religion')
    plot_word_frequencies(president, political_count, 'Political')
    plot_word_frequencies(president, socio_cultural_count, 'Socio-cultural')
    print("\n")
```

### Presidents' word frequencies on the same graph

```python
"""Here, I plotted all the presidents' word frequencies on the same graph using multiple bar charts (also called 
    grouped bar charts). Each president's set of bars will represent their religious, political, or 
    socio-cultural word frequencies, and a legend will distinguish which bars correspond to which president."""


import numpy as np  #with the help of CHATGPT; I realized I needed to import numpy before I could achieve my goal
import matplotlib.pyplot as plt

# Function to plot combined word frequencies for all presidents, with space between clusters
def plot_combined_word_frequencies(presidents, theme_word_frequencies, theme_name):
    """Plots the frequency of words for a given theme in speeches from multiple presidents (horizontal bars with space between clusters)."""
    
    words = list(theme_word_frequencies[list(presidents.keys())[0]].keys())  # Words in the theme
    num_words = len(words)
    
    # Create indices with space between clusters of bars (increase the gap between clusters)
    gap = 2  # Defines the gap between clusters of presidents
    indices = np.arange(0, gap * len(presidents), gap)  # Larger gap between presidents
    
    # Bar height (small enough to fit all bars side by side)
    bar_height = 0.2
    
    plt.figure(figsize=(10, 6))
    
    # Plot bars for each word
    for i, word in enumerate(words):
        # Collect frequencies for each president
        freqs = []  # Reset the list for each word
        for president in presidents:
            if word in theme_word_frequencies[president]:
                freqs.append(theme_word_frequencies[president][word])
            else:
                freqs.append(0)
        
        # Plot bars for each word
        plt.barh(indices + i * bar_height, freqs, bar_height, label=word)
    
    # Add labels and title
    plt.xlabel('Frequency')
    plt.ylabel('Presidents')
    plt.title(f'{theme_name} Theme Word Frequencies Across Presidents')
    plt.yticks(indices + bar_height * (num_words - 1) / 2, presidents.keys())  # Set president names on y-axis
    plt.legend(title="Words")
    plt.show()

# Collect frequencies for each theme for all presidents
religion_word_frequencies = {}
political_word_frequencies = {}
socio_cultural_word_frequencies = {}

# Loop through each president's speech and collect word frequencies
for president, speech in presidents.items():
    # Process the speech
    tokens = process_speeches(speech)
    
    # Count word frequencies for each theme
    religion_word_frequencies[president] = count_theme_words(tokens, religion_words)
    political_word_frequencies[president] = count_theme_words(tokens, political_words)
    socio_cultural_word_frequencies[president] = count_theme_words(tokens, socio_cultural_words)

# Plot combined word frequencies for religion, political, and socio-cultural themes with presidents on the y-axis
plot_combined_word_frequencies(presidents, religion_word_frequencies, 'Religion')
plot_combined_word_frequencies(presidents, political_word_frequencies, 'Political')
plot_combined_word_frequencies(presidents, socio_cultural_word_frequencies, 'Socio-cultural')
```

### ConditionalFreqDist for analysing economic-related words

```python
"""I also looked at words like "economy," "jobs," "market," "growth," and "crisis" to explore how different 
presidents have focused on economic issues, particularly during recessions or periods of economic growth."""

# Define the economic-related words to analyze
economic_words = ['economy', 'jobs', 'market', 'growth', 'crisis']

# Creating a ConditionalFreqDist for economic-related words
plt.figure(figsize=(20, 10))
cfd_economic = nltk.ConditionalFreqDist((target, fileid[:4])
                                        for fileid in inaugural.fileids()
                                        for word in inaugural.words(fileid)
                                        for target in economic_words
                                        if word.lower().startswith(target))

# Plot the frequency distribution of economic-related words over time
cfd_economic.plot(title="Economic Themes in US Presidential Speeches Over Time")
```

### Analyzing word pairs (bigrams) 

```python
""" I looked beyond analyzing individuals words to analyzing word pairs (bigrams)
    to see which combinations of words are most frequently used together. This will reveal the most common 
    expression used by these presidents frequently """

#importing the bigrams package
from nltk import bigrams, FreqDist

# Creating an empty list to store all the words from president's speeches
all_words = []

# Looping through each inaugural file, tokenize the text, and filter out stop words
for fileid in inaugural.fileids():
    speech_text = word_tokenize(inaugural.raw(fileid))
    for word in speech_text:
        # Only keep alphabetic words and remove stopwords
        if word.isalpha() and word.lower() not in stop_words:
            all_words.append(word.lower())

# Generate bigrams from the tokenized words (after removing stopwords)
bigrams_list = list(bigrams(all_words))

# Calculate frequency distribution of bigrams
fdist_bigrams = FreqDist(bigrams_list)

# Plot the top 20 most common bigrams
plt.figure(figsize=(12, 6))
fdist_bigrams.plot(20, title="Top 20 Most Common Bigrams in U.S. Presidential Speeches (Stop Words Removed)")
```

### Comparing Reagan's first and second inaugural speeches to identify shifts or consistencies

```python
"""This code compares Reagan's first and second inaugural speeches to identify shifts or consistencies in his 
    rhetoric and focus over his two terms. By analyzing word frequencies, it highlights changes in key themes 
    and priorities, offering insights into how Reagan's messaging evolved in response to political or social 
    developments during his president"""

# Compare first and second inaugural speeches of specific presidents
first_speech = inaugural.raw('1981-Reagan.txt')
second_speech = inaugural.raw('1985-Reagan.txt')

# Tokenize and filter out stop words for the first speech
first_words = []
for word in word_tokenize(first_speech):
    if word.isalpha() and word.lower() not in stop_words:
        first_words.append(word.lower())

# Tokenize and filter out stop words for the second speech
second_words = []
for word in word_tokenize(second_speech):
    if word.isalpha() and word.lower() not in stop_words:
        second_words.append(word.lower())

# Calculate word frequency distribution for both speeches
first_words_freq = nltk.FreqDist(first_words)
second_words_freq = nltk.FreqDist(second_words)

# Plot and compare the word frequencies for the first speech
plt.figure(figsize=(12, 6))
first_words_freq.plot(20, title="Most Common Words in Reagan's First Inaugural Speech (Stop Words Removed)")

# Plot and compare the word frequencies for the second speech
plt.figure(figsize=(12, 6))
second_words_freq.plot(20, title="Most Common Words in Reagan's Second Inaugural Speech (Stop Words Removed)")
```

### Tracking the usage of the word 'Freedom" over time

```python
# I tracked the usage of the word 'Freedom" over time
# Define the word to track (e.g., "freedom")
word_to_track = 'freedom'

# Loop through speeches, tracking the word's frequency over time
frequencies = []
years = []
for fileid in inaugural.fileids():
    words = [word.lower() for word in inaugural.words(fileid) if word.isalpha()]
    freq = words.count(word_to_track)
    frequencies.append(freq)
    years.append(fileid[:4])

# Plot the frequency of the word over time
plt.figure(figsize=(12, 6))
plt.plot(years, frequencies, marker='o')
plt.title(f'Usage of the word "{word_to_track}" in U.S. Presidential Speeches Over Time')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Calculating the lexical diversity to reveal the richness and variety of vocabulary of Top 5 Presidents

```python
"""I calculated the lexical diversity to reveal the richness and variety of vocabulary used by different 
    presidents. It will provide insight into their rhetorical style—whether a speech is more sophisticated and 
    diverse or simple and repetitive."""

def lexical_diversity(text):
    """Function to calculate lexical diversity of the presidents """
    return len(set(text)) / len(text) #formular for calculating lexical diversity (Number of unique words/Total number of words)

# Storing the results of lexical diveristy for each president's speech
diversity_scores = []

# Loop through a selection of speeches and calculate lexical diversity
for fileid in inaugural.fileids()[:]:  # this will loop through all the president
    words = []
    # Loop through each word in the speech and filter alphabetic words
    for word in inaugural.words(fileid):
        if word.isalpha():
            words.append(word.lower())
    
    # Calculate lexical diversity and store the result along with the president's fileid
    diversity = lexical_diversity(words)
    diversity_scores.append((fileid, diversity))  # Store tuple of fileid and lexical diversity

# Sort the diversity scores in descending order based on the diversity score
diversity_scores.sort(key=lambda x: x[1], reverse=True)

# Print the top 5 presidents with the highest lexical diversity
print("\nTop 5 Presidents with Highest Lexical Diversity:\n")
for fileid, diversity in diversity_scores[:5]:
    print(f"{fileid}: {diversity:.4f}")
```

### Intro to Summarization
Summarization has been a thorny problem for data scientists over the past decade. There have been many approaches, but most, until ChatGPT and it's use of the transformers' architecture, they have been unsatisfactory. There are two main types of summarization:

Extractive Summarization: Finding the most important sentences and rendering them verbatim.

Abstractive Summarization: Taking the most important ideas from a document and generating a summary based on them

There have been many different algorithms that have attempted summarization. Let's learn about one of the most used and deploy it in python.

#### Luhns Heuristic Method
This is one of the earliest approaches of text summarization. This is related to our manual interpretation of the importance of word frequency in a document.Basically, just as we did, Luhn recognized that word frequency and word significance is highly correlated, but only "important" words. The assumption that the most important sentences are those with the most significant words. Here is a nice short article that explains the algo in an accessible way: https://blog.fastforwardlabs.com/2016/03/25/h.p.-luhn-and-the-heuristic-value-of-simplicity.html
